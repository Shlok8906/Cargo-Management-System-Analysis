```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=6)
df_imputed = imputer.fit_transform(df.select_dtypes(include=['float64','int64']))
df[df.select_dtypes(include=['float64','int64']).columns] = df_imputed
df.head()
df.to_csv('imputed_dataset.csv', index=False)
```

#EDA SECTION


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.select_dtypes(include=['float64','int64']))
df[df.select_dtypes(include=['float64','int64']).columns] = df_scaled
```

```python
# --- Chart 1: Scatter Plot (Cost vs. Distance) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Distance_km', y='Cost_USD', hue='Transport_Method', data=df)
plt.title('Shipment Cost vs. Distance by Transport Method')
plt.xlabel('Distance (km)')
plt.ylabel('Cost (USD)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Priority', y='Transit_Time_Days', data=df)  # use df, not df_scaled
plt.title('Transit Time vs. Shipment Priority')
plt.xlabel('Priority')
plt.ylabel('Transit Time (Days)')
plt.show()
```

```python
# Visualize the distribution of shipment transit times
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Transit_Time_Days', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Shipment Transit Times')
plt.xlabel('Transit Time (Days)')
plt.ylabel('Number of Shipments')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Status', data=df, order=df['Status'].value_counts().index, palette='viridis')
plt.title('Total Shipments by Status')
plt.xlabel('Status')
plt.ylabel('Number of Shipments')
plt.show()
```

```python
plt.figure(figsize=(10,7))
numerical_cols = ['Weight_kg', 'Volume_m3', 'Cost_USD', 'Transit_Time_Days', 'Distance_km', 'Fuel_Consumed_L', 'CO2_Emissions_kg', 'Cargo_Value_USD']
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Cargo_Type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set2')
plt.title('Distribution of Cargo Types')
plt.ylabel('')
plt.show()
```

```python
carrier_status_counts = df.groupby(['Carrier', 'Status']).size().unstack(fill_value=0)
carrier_status_counts.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='tab20')
plt.title('Shipment Status by Carrier')
plt.xlabel('Carrier')
plt.ylabel('Number of Shipments')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Status')
plt.show()
```

```python
# 1. Top 10 Origins
plt.figure(figsize=(10,6))
df['Origin'].value_counts().head(10).plot(kind='bar', color="skyblue")
plt.title("Top 10 Shipment Origins")
plt.ylabel("Count")
plt.xlabel("Origin")
plt.show()
```

```python
# 2. Top 10 Destinations
plt.figure(figsize=(10,6))
df['Destination'].value_counts().head(10).plot(kind='bar', color="salmon")
plt.title("Top 10 Shipment Destinations")
plt.ylabel("Count")
plt.xlabel("Destination")
plt.show()
```

```python
# 4. Transport Method vs Average Cost
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Transport_Method", y="Cost_USD", estimator=np.mean, palette="Set1")
plt.title("Average Cost by Transport Method")
plt.ylabel("Average Cost (USD)")
plt.show()
```

```python
# 5. Weight vs Cargo Type
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Cargo_Type", y="Weight_kg", palette="Set3")
plt.xticks(rotation=45)
plt.title("Weight Distribution by Cargo Type")
plt.show()
```

```python
# 4. Average Cost by Cargo Type
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Cargo_Type", y="Cost_USD", estimator=np.mean, palette="Set3")
plt.xticks(rotation=45)
plt.title("Average Cost by Cargo Type")
plt.ylabel("Average Cost (USD)")
plt.show()
```

```python
# 5. Average Cost by Priority
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Priority", y="Cost_USD", estimator=np.mean, palette="Set1")
plt.title("Average Cost by Priority Level")
plt.ylabel("Average Cost (USD)")
plt.show()
```

```python
# 7. Top 10 Expensive Origins
top_origins = df.groupby("Origin")["Cost_USD"].mean().nlargest(10).index
plt.figure(figsize=(10,6))
sns.barplot(data=df[df["Origin"].isin(top_origins)], x="Origin", y="Cost_USD", estimator=np.mean, palette="coolwarm")
plt.title("Top 10 Origins with Highest Average Shipment Cost")
plt.ylabel("Average Cost (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Cost Breakdown (Customs + Insurance + Base Cost)
df["Total_Extra_Cost"] = df["Customs_Fees_USD"] + df["Insurance_Value_USD"]
breakdown = df.groupby("Transport_Method")[["Cost_USD", "Customs_Fees_USD", "Insurance_Value_USD"]].mean()

breakdown.plot(kind="bar", stacked=True, figsize=(10,6), color=["skyblue","orange","green"])
plt.title("Average Cost Breakdown by Transport Method")
plt.ylabel("Average Cost (USD)")
plt.xlabel("Transport Method")
plt.legend(["Base Cost", "Customs Fees", "Insurance Value"])
plt.show()
```
